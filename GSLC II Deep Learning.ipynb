```python
df = pd.read_csv('drive/MyDrive/Sentiment-analysis/train.csv', encoding = 'ISO-8859-1')
df
```

```python
from sklearn.model_selection import train_test_split

df_train, df_val = train_test_split(df, test_size = 0.2, random_state = 42)
```

```python
df_test = pd.read_csv('drive/MyDrive/Sentiment-analysis/test.csv', encoding = 'ISO-8859-1')
df_test
```

```python
df_train = df_train.dropna()
df_val = df_val.dropna()
df_test = df_test.dropna()
```

```python
x_train = df_train['text'].values
y_train = df_train['sentiment'].values
x_val = df_val['text'].values
y_val = df_val['sentiment'].values
x_test = df_test['text'].values
y_test = df_test['sentiment'].values
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words = 10000)
tokenizer.fit_on_texts(x_train)

x_train_seq = tokenizer.texts_to_sequences(x_train)
x_train_pad = pad_sequences(x_train_seq, padding = "post", truncating = "post")

x_val_seq = tokenizer.texts_to_sequences(x_val)
x_val_pad = pad_sequences(x_val_seq, padding = "post", truncating = "post")

x_test_seq = tokenizer.texts_to_sequences(x_test)
x_test_pad = pad_sequences(x_test_seq, padding = "post", truncating = "post")
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.layers import CategoryEncoding

encoded_y_train = LabelEncoder().fit_transform(df_train['sentiment'])
encoded_y_train = CategoryEncoding(num_tokens = 3, output_mode = "one_hot")(encoded_y_train)
print(encoded_y_train)
```

```python
encoded_y_val = LabelEncoder().fit_transform(df_val['sentiment'])
encoded_y_val = CategoryEncoding(num_tokens = 3, output_mode = "one_hot")(encoded_y_val)
print(encoded_y_val)
```

```python
encoded_y_test = LabelEncoder().fit_transform(df_test['sentiment'])
encoded_y_test = CategoryEncoding(num_tokens = 3, output_mode = "one_hot")(encoded_y_test)
print(encoded_y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, SimpleRNN, Dropout, Flatten, Bidirectional, LSTM, GRU

model = Sequential([
    Embedding(output_dim = 10000, input_dim = 128, mask_zero = True),
    GRU(64, return_sequences = True),
    GRU(64),
    Dense(128, activation = "relu"),
    Dropout(0.5),
    Dense(3, activation = 'softmax')
])
```

```python
print(model.summary())
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer = Adam(learning_rate = 0.001), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
train_model = model.fit(x_train_pad, encoded_y_train, validation_data = (x_val_pad, encoded_y_val), batch_size = 32, epochs = 30)
```

```python
plt.plot(train_model.history['accuracy'])
plt.plot(train_model.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'validation'])
plt.show()
```

```python
plt.plot(train_model.history['loss'])
plt.plot(train_model.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'validation'])
plt.show()
```

```python
eval_model = model.evaluate(x_test_pad, encoded_y_test)
print(eval_model)
```

```python
from sklearn.metrics import f1_score, roc_auc_score

pred = list(model.predict(x_test_pad))
pred = [[int(col == max(row)) for col in row] for row in pred]

# print((encoded_y_test.ref()), (pred), sep = "\n")

print(f"F1 score: {f1_score(encoded_y_test, pred, average = 'micro') :.4f}")
print(f"AUC score: {roc_auc_score(encoded_y_test, pred, multi_class = 'ovo', average = 'micro') :.4f}")
```
